Since there's an obvious right skew in our data, we must deal with this before moving on to the analysis. To have a better understanding of the extreme values we will use _modified z-scores_ (computed using the median) and observe the extreme values up close using this method. We won't be using standard z-scores in this case because the mean and standard deviation are significantly more sensitive to the presence of anomalies compared to the median. Standard z-scores work best with a Gaussian (normal) distribtuion, hence for heavily skewed data, this modified version will be more suitable. Since there's no universal rule for a threshold for MZS, we will use a threshold of +3 to detect outliers as commonly used. 

The _Modified Z-Score_ (MZS) is given by the formula:

$$
\text{Modified Z-Score} = \frac{0.6745 (x_i - \hat{x})}{\text{MAD}}
$$

where:
- $x_i$ is an individual data point.
- $\hat{x}$ is the median of the dataset.
- MAD stands for the Median Absolute Deviation, calculated as:

  $$
  \text{MAD} = \text{median}(\left| x_i - \text{median}(X) \right|)
  $$

where:
- $X$ represents the entire dataset.



The next steps:

1. Compute the modified z-scores using (x - median(x))/mad(x)) where mad will be calculated using the `mad()` function
2. Create a new variable with the modified z-scores called `mod_zscore`.
3. Create a new dataset with the observations that can be considered outliers (mod_zscore > 3).
4. Find the proportion of extreme values in the dataset by dividing the outliers by the total number of observations. 


```r
# Set the threshold
mad_threshold <- 6.5

# Total number of observations in the data
n <- nrow(news_data)

# Calculate Median and MAD
median_shares <- median(news_data$shares, na.rm = TRUE)
mad_shares <- mad(news_data$shares, constant = 1, na.rm = TRUE)

# Calculate the Modified Z-Score
news_data$mod_zscore <- (news_data$shares - median_shares) / mad_shares

# Add an id numbers for each observations for anti_join which is for further operations
news_data$n <- c(1:n)

# Calcualte the portion of the data that can be considered outliers with the set threshold of 3
outliers <- news_data %>% filter(news_data$mod_zscore > mad_threshold | news_data$mod_zscore < -mad_threshold)
non_outliers <- anti_join(x = news_data, y = outliers, by = "n")

# Count the number of outliers
n_outliers <- nrow(outliers)

cat("Proportion of values that can be considered outliers:")
# Calculate the ratio
ratio_outliers <- n_outliers / n
ratio_outliers

slice_min(.data = outliers, n = 10, order_by = mod_zscore)

outliers_summary <- non_outliers %>%
    group_by(channel_type) %>%
    summarise(n = n(),
              mean = mean(shares),
              min = min(shares),
              max = max(shares))
outliers_summary
######## learn if dividing is smart, then see how you should devide!!!!!
```
